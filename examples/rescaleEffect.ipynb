```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("../data_test/SRP109549.tsv", sep='\t', index_col=0)
df_scaled = pd.read_csv("../data_test/SRP109549_rescaled.tsv", sep='\t', index_col=0)
```

```python
scaler= MinMaxScaler((0,20))
df_sample_rescaled = scaler.fit_transform(df.T)
df_sample_rescaled = pd.DataFrame(df_sample_rescaled.T, index=df.index, columns=df.columns)
```

```python
# random_gene = np.random.choice(df.columns,1)[0]
random_gene = 'MYCN'
```

```python
random_gene
```

```python
df[random_gene].hist(bins=10)
plt.title('Original DF')
plt.show()
plt.close()
```

```python
df_scaled[random_gene].hist(bins=10)
plt.title('Gene Scaled DF')
plt.show()
plt.close()
```

```python
df_sample_rescaled[random_gene].hist(bins=10)
plt.title('Samples Scaled DF')
plt.show()
plt.close()
```
